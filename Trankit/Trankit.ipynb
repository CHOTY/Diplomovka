```python
from trankit import Pipeline
import pandas as pd
```

```python
def read_text_file(filepath):
    with open(filepath, "r", encoding="utf-8") as file:
        text = file.read()
    return text

filepath = "DiploDiktaty.txt"
text = read_text_file(filepath)
```

```python
# Inicializácia pipeline pre slovenský jazyk
pipeline = Pipeline(lang="slovak", cache_dir="./trankit_cache", gpu=False)  # Nastavenie na CPU
```

```python
doc = pipeline.posdep(text)
```

```python
doc
```

```python
# Inicializácia zoznamu pre uloženie dát
words_data = []

# Prechádzanie cez vety a ich tokeny
for sentence in doc['sentences']:
    sentence_id = sentence['id']
    for token in sentence['tokens']:
        word_id = token['id']
        word_text = token['text']
        xpos = token['xpos']

        # Rozdelenie XPOS na jednotlivé časti
        xpos1 = xpos[0] if len(xpos) > 0 else '-'
        xpos2 = xpos[1] if len(xpos) > 1 else '-'
        xpos3 = xpos[2] if len(xpos) > 2 else '-'
        xpos4 = xpos[3] if len(xpos) > 3 else '-'
        xpos5 = xpos[4] if len(xpos) > 4 else '-'
        xpos6 = xpos[5] if len(xpos) > 5 else '-'
        xpos7 = xpos[6] if len(xpos) > 6 else '-'

        # Pridanie dát do zoznamu
        words_data.append([word_text, xpos1, xpos2, xpos3, xpos4, xpos5, xpos6, xpos7, word_id])

# Vytvorenie DataFrame
words_df = pd.DataFrame(words_data, columns=['Token', 'XPOS1', 'XPOS2', 'XPOS3', 'XPOS4', 'XPOS5', 'XPOS6', 'XPOS7', 'Word_ID'])

# Uloženie do Excel
words_df.to_excel("words.xlsx", index=False, engine="openpyxl")

print("Výstup uložený do 'words.xlsx'.")
```
