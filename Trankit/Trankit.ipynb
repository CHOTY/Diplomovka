```python
import stanza
import pandas as pd
from trankit import Pipeline
```

```python
nlp = stanza.Pipeline(lang='sk', processors='tokenize,pos', use_gpu=False)

# Funkcia na načítanie textového súboru
def read_text_file(filepath):
    with open(filepath, "r", encoding="utf-8") as file:
        return file.read()
```

```python
# Načítanie textu
filepath = "DiploDiktaty.txt"
text = read_text_file(filepath)

# Spracovanie textu (tokenizácia a ďalšie, ak sú potrebné)
doc = nlp(text)

# Inicializácia zoznamu pre ukladanie dát
data = []

# Inicializujeme vlastný čítač pre Sentence_ID a Word_ID
sentence_counter = 1
word_counter = 1

# Prechádzame cez vety a tokeny v dokumente
for sentence in doc.sentences:
    current_sentence_id = sentence_counter  # generujeme vlastné Sentence_ID
    sentence_counter += 1
    for token in sentence.tokens:
        # Vygenerujeme vlastné Word_ID (kumulatívne pre celý dokument)
        current_word_id = word_counter
        word_counter += 1

        # Uložíme do zoznamu: Sentence_ID, Word_ID, Token text
        data.append([current_sentence_id, current_word_id, token.text])

# Vytvorenie DataFrame so stĺpcami: Sentence_ID, Word_ID, Token
df = pd.DataFrame(data, columns=['Sentence_ID', 'Word_ID', 'Token'])
df
```

```python
# Predpokladajme, že máš DataFrame 'df' s tokenmi (napr. zo Stanzovho tokenizovania)
# a stĺpec 'Token', potom vytvoríš zoznam "jednoslovných viet"
single_token_sentences = df['Token'].apply(lambda token: token).tolist()

# Teraz môžeš tieto "vety" (každá obsahuje len jeden token) spojiť do textu, kde každý riadok je jedna veta:
text_for_trankit = "\n".join(single_token_sentences)

# Inicializácia Trankit pipeline
trankit_pipeline = Pipeline(lang="slovak", cache_dir="./trankit_cache", gpu=False)

# Spracovanie textu
result = trankit_pipeline.posdep(text_for_trankit)
print(result)
```

```python
# Inicializácia zoznamu pre uloženie dát
words_data = []

# Prechádzanie cez vety a ich tokeny
for sentence in result['sentences']:
    for token in sentence['tokens']:
        word_id = sentence['id']         # Získame Word_ID
        word_text = token['text']
        xpos = token['xpos']

        # Rozdelenie XPOS na jednotlivé časti
        xpos1 = xpos[0] if len(xpos) > 0 else '-'
        xpos2 = xpos[1] if len(xpos) > 1 else '-'
        xpos3 = xpos[2] if len(xpos) > 2 else '-'
        xpos4 = xpos[3] if len(xpos) > 3 else '-'
        xpos5 = xpos[4] if len(xpos) > 4 else '-'
        xpos6 = xpos[5] if len(xpos) > 5 else '-'
        xpos7 = xpos[6] if len(xpos) > 6 else '-'


        # Pridanie dát do zoznamu
        words_data.append([word_text, xpos1, xpos2, xpos3, xpos4, xpos5, xpos6, xpos7, word_id])

# Vytvorenie DataFrame
words_df = pd.DataFrame(words_data, columns=['Token', 'Xpos1', 'Xpos2', 'Xpos3', 'Xpos4', 'Xpos5', 'Xpos6', 'Xpos7', 'Word_ID'])

# Uloženie do Excel
words_df.to_excel("words.xlsx", index=False, engine="openpyxl")

print("Výstup uložený do 'words.xlsx'.")
```
