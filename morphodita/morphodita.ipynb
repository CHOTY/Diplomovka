```python
from corpy.morphodita import Tagger
from corpy.morphodita import Tokenizer
import pandas as pd
```

```python

# Načítanie textového súboru
def read_text_file(filepath):
    with open(filepath, "r", encoding="utf-8") as file:
        return file.read()

# Cesta k textovému súboru
filepath = r"C:\Users\marti\Downloads\DiploDiktaty.txt"
text = read_text_file(filepath)

# Cesta k MorphoDiTa modelu
model_path = r"slovak-morfflex-pdt-170914/slovak-morfflex-pdt-170914.tagger"

# Inicializácia MorphoDiTa Tagger-u cez Corpy
tagger = Tagger(model_path)

# Tokenizácia a tagovanie textu
sentences_data = []  # Dáta pre sentences.csv
words_data = []      # Dáta pre words.csv

# Tagovanie viet
sents = tagger.tag(text, sents=True)  # Vrátia sa vety s tokenmi a ich informáciami

# Spracovanie viet a slov
for sentence_id, sentence in enumerate(sents, start=1):
    sentence_text = " ".join(token.word for token in sentence)  # Získanie textu vety
    sentences_data.append([sentence_id, sentence_text])  # Uloženie vety do zoznamu
    print(sentence)
    for token in sentence:
        words_data.append([
            sentence_id,    # ID vety
            token.word,     # Token (slovo)
            token.lemma,    # Lemma
            token.tag       # Morfologický tag
        ])

# Uloženie výsledkov do CSV
sentences_df = pd.DataFrame(sentences_data, columns=["Sentence_ID", "Text"])
words_df = pd.DataFrame(words_data, columns=["Sentence_ID", "Token", "Lemma", "Tag"])

sentences_df.to_excel("sentences.xlsx", index=False, engine="openpyxl")
words_df.to_excel("words.xlsx", index=False, engine="openpyxl")

print("Výstup uložený do 'sentences.csv' a 'words.csv'.")
```
